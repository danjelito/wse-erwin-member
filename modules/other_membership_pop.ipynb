# Other Membership Population

```python
(df
    .assign(
        product= (df['addon_1'].fillna('') + ' ' + 
                  df['addon_2'].fillna('') + ' ' + 
                  df['addon_3'].fillna('')
                  ).str.strip()
    )
    .groupby('product')
    .sum(numeric_only= True)
    # filer column
    .filter(regex= ("active.+"))
    # filter rows
    # filter out ST
    .loc[(lambda df_: ~(df_.index.isin([' ', '', 'Street Talk', 'Other'])))]
    # clean index name
    .reset_index()
    # rename columns
    .rename(columns= lambda c: c.replace('_', ' ').replace('active ', '').title())
    # transpose
    .set_index('Product')
    .transpose()
 )
```
