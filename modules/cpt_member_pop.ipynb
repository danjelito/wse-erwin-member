# CPT Population

```python
(df
    # filter for cpt
    .loc[df['is_cpt'] == True]
    # group by core membership
    .groupby('core_product')
    .sum(numeric_only= True)
    # filer column
    .filter(regex= ("active.+"))
    # clean index name
    .rename(columns= lambda c: c.replace('_', ' ').replace('active ', '').title())
    .transpose()
    .rename_axis('Month', axis= 1)
    # drop where sum = 0
    .loc[:, lambda df_: df_.sum() > 0]
)
```
